```python
# https://leetcode.com/contest/weekly-contest-257/problems/count-special-quadruplets/
from itertools import combinations
def countQuadruplets(nums):
    result = 0
    idx = list(range(0, len(nums)))
    comb = combinations(idx, 3)
    for c in comb:
        diff = set(idx) - set(c)
        sum_ary = [nums[v] for v in c]
        res_ary = [nums[x] for x in diff]
        print(sum_ary, res_ary)
        total = sum(sum_ary)
        if total in res_ary:
            result += 1
    return result
#print(countQuadruplets(nums = [1,2,3,6]))
#print(countQuadruplets(nums = [3,3,6,4,5]))
#print(countQuadruplets(nums = [1,1,1,3,5]))
#print(countQuadruplets(nums = [9,6,8,23,39,23]))
print(countQuadruplets(nums = [28,8,49,85,37,90,20,8]))
```
